# Text Summarization

## [TextRank Algorithm](https://cran.r-project.org/web/packages/textrank/vignettes/textrank.html)
1. Split the document into sentences
2. Get a sentence embedding for each sentence
3. Build a graph where nodes are the sentences and edge weights are the similarities (e.g. cosine similarity) of the sentence embeddings.
4. Run the PageRank algorithm on the graph to obtain a PageRank score for each sentence. A high PageRank score means that the node is important for the network.
5. Keep the sentences with a score higher than a certain threshold to use them as the summary of the document.

[Python repo](https://github.com/DerwenAI/pytextrank)

## Sentence Embedding Clustering

### Coreference Resolution
Coreference Resolution has the goal of finding, grouping, and then substituting an ambiguous expressions with the real-world entities they are referring to.

### [BERT Extractive Summarizer](https://github.com/dmmiller612/bert-extractive-summarizer)
1. Resolve coreferences in the document
2. Split the document into sentences
3. Get a sentence embedding (e.g. using BERT) for each sentence
4. Run K-Means on the sentence embedding and get K clusters. K will be the number of sentences in the summary
5. Find the closest sentence to the centroid of each cluster and use them to compose the summary

## Seq-to-Seq Encoder-Decoder Modelling
1. Get a dataset with documents paired with summaries
2. Choose a suitable metric for text summarization (e.g., [ROUGE](https://en.wikipedia.org/wiki/ROUGE_(metric)))
3. Train a seq-to-seq model (usually a trasnformer-based model) to produce summaries from texts in a supervised way.

## Proposal: Hybrid Extractive-Abstractive System
1. Coreference resolution to retain context between sentences
2. Use TextRank to filter the most common/important information from the document.
3. Parse it through a pre-trained encoder-decoder (Transformer, RNN, LSTM)

Since the TextRank algorithm splits the document into sentences, context from a sentence may be lost if it refers to an entity in the previous sentence:  
> “This software license agreement governs your access to and use of the application and shall remain in effect unless terminated in accordance with these terms.”  
> "It shall be terminated upon thirty days notice."  

**Problem:** without the previous sentence, the system would have no context as to what "It" refers to.  

By condensing the input sequence, it reduces the need for a larger context window.  
This way we can reduce the vanishing gradient problem in RNN or LSTM and reduce hallucination in Transformer architectures.
