```python
# examine list off top-ranked phrases in the document
for p in doc._.phrases:
    ic(p.rank, p.count, p.text)
    ic(p.chunks)
```

```python
# contruct a list of sentence boundaries with a phrase vector (initialized to empty set) for each
sent_bounds = [ [s.start, s.end, set([])] for s in doc.sents ]
sent_bounds
```

```python
# Iterate through the top-ranked phrases, added them to the phfrase vector for each sentence

limit_phrases = 10

phrase_id = 0
unit_vector = []

for p in doc._.phrases:
    ic(phrase_id, p.text, p.rank)
    
    unit_vector.append(p.rank)
    
    for chunk in p.chunks:
        ic(chunk.start, chunk.end)
        
        for sent_start, sent_end, sent_vector in sent_bounds:
            if chunk.start >= sent_start and chunk.end <= sent_end:
                ic(sent_start, chunk.start, chunk.end, sent_end)
                sent_vector.add(phrase_id)
                break

    phrase_id += 1

    if phrase_id == limit_phrases:
        break
```

```python
for sent in doc.sents:
    ic(sent)
```

```python
sum_ranks = sum(unit_vector)

unit_vector = [rank/sum_ranks for rank in unit_vector]
unit_vector
```

```python
# Iterate through each sentece, calculating its euclidean distance from the unit vector
from math import sqrt

sent_rank = {}
sent_id = 0

for sent_start, sent_end, sent_vector in sent_bounds:
    ic(sent_vector)
    sum_sq = 0.0
    ic

    for phrase_id in range(len(unit_vector)):
        ic(phrase_id, unit_vector[phrase_id])

        if phrase_id not in sent_vector:
            sum_sq += unit_vector[phrase_id]**2

        sent_rank[sent_id] = sqrt(sum_sq)
        sent_id += 1
```

```python
limit_sentences = 2

sent_text = {}
sent_id = 0

for sent in doc.sents:
    sent_text[sent_id] = sent.text
    sent_id += 1

num_sent = 0

for sent_id, rank in sorted(sent_rank.items()):
    ic(sent_id, sent_text[sent_id])
    num_sent += 1
    
    if num_sent == limit_sentences:
        break
```
